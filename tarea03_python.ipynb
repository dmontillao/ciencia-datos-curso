13. Función contar_python
Cree una función llamada contar_python que cuente la cantidad de veces que aparece la palabra "python" en una cadena (ignorando mayúsculas y minúsculas).

```python
def contar_python(texto):
    convertir_texto = texto.lower()
    return convertir_texto.count("python")
```

```python
contar_python("Python es un lenguaje de programacion, python es divertido")
```

14. Un científico de datos necesita desarrollar una función para detectar anomalías en un conjunto de datos de valores positivos. La función deteccion_anomalia debe seguir estas reglas:

Recibe los siguientes argumentos:

valor (float): El valor a evaluar.

valor_media (float): La media del conjunto de datos.

desviacion_estandar (float): La desviación estándar.

distribucion_normal (bool): Indica si la distribución es normal.

Si la distribución es normal y el valor es mayor a valor_media + 3 * desviacion_estandar, la función retorna "anormal".

Si la distribución es normal y el valor es menor a valor_media - 3 * desviacion_estandar, la función retorna "normal".

En cualquier otro caso, retorna "indeterminado".

```python
def deteccion_anomalia(valor, valor_media, desviacion_estandar, distribucion_normal):
  if distribucion_normal and (valor < valor_media - 3 * desviacion_estandar):
     return "normal"
  elif distribucion_normal and (valor > valor_media + 3 * desviacion_estandar):
    return "anomalia"
  else:
    return "indeterminado"

```

```python
print(deteccion_anomalia(
    valor=45.4, valor_media=9.5, desviacion_estandar=8.7, distribucion_normal=True
))  # Se espera: 'anormal'
```
