¿Cuántos puestos de trabajo Unicos existieron en 2013? (Ejemplo, los títulos de trabajo con solo una ocurrencia en 2013?)

```python
df_2013 = df[df["Year"] == 2013]
puestos_unicos = df_2013.groupby('JobTitle').filter(lambda x: len(x) == 1)
cantidad_puestos_unicos = puestos_unicos['JobTitle'].nunique()
print(cantidad_puestos_unicos)
```

¿Cuántas personas tienen la palabra Chief en el título de su trabajo? Apoyese en los métodos para manipular strings y en el método .apply()

```python
def tiene_chief(titulo):
    return 'chief' in str(titulo).lower()

mask_chief = df['JobTitle'].apply(tiene_chief)

cantidad_chief = mask_chief.sum()

print(cantidad_chief)
```

Bono: ¿Existe una correlación entre la longitud del nombre del Título del trabajo y el Salario?

crea una nueva columna donde se guarde la longitud del titulo del trabajo. tip: usar el método .apply()

```python
df['LongitudTitulo'] = df['JobTitle'].apply(len)

print(df[['JobTitle', 'LongitudTitulo', 'TotalPayBenefits']].head())
```

```python
calcu_correlacion = df['LongitudTitulo'].corr(df['TotalPayBenefits'])
print(f"La correlación es: {calcu_correlacion}")
print(df[['LongitudTitulo','TotalPayBenefits']].head())
```
